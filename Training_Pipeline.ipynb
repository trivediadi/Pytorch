# Simple neural network

```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.drop(['id','Unnamed: 32'],axis=1,inplace=True)
```

```python
df
```

# train test split

```python
X_train,X_test,y_train,y_test=train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size=0.2,random_state=42)
```

# Scaling

```python
scalar=StandardScaler()
X_train=scalar.fit_transform(X_train)
X_test=scalar.fit_transform(X_test)
```

```python
X_train.shape
```

```python
y_train
```

#Label Encoder

```python
encoder=LabelEncoder()
y_train=encoder.fit_transform(y_train)
y_test=encoder.fit_transform(y_test)
```

```python
# B=0,M=1
y_train
```

#Numpy array to Pytorch tensor



```python
X_train_tensor=torch.from_numpy(X_train)
X_test_tensor=torch.from_numpy(X_test)
y_train_tensor=torch.from_numpy(y_train)
y_test_tensor=torch.from_numpy(y_test)
```

```python
X_train_tensor.shape
```

# Model Defining

```python
class SimpleNN():
  def __init__(self,inputs,neurons):
    self.weights=torch.rand(inputs,1,dtype=torch.float64,requires_grad=True)
    self.bias=torch.zeros(neurons,dtype=torch.float64,requires_grad=True)
  def forward(self,input):
    z=torch.matmul(input,self.weights)+self.bias
    y_pred=torch.sigmoid(z)
    return y_pred
  def loss_function(self,y_true,y_pred):
    epsilion=1e-8
    y_pred=torch.clamp(y_pred,epsilion,1-epsilion)
    return -(y_true*torch.log(y_pred)+(1-y_true)*torch.log(1-y_pred)).mean()
```

#Important parameters

```python
learning_rate=0.1
epochs=25
```

#Training Pipeline

```python
#create Model
model=SimpleNN(X_test_tensor.shape[1],1)
for epoch in range(epochs):
  #forward pass
  y_pred=model.forward(X_train_tensor)
  #loss calculation
  loss=model.loss_function(y_train_tensor,y_pred)
  #backward pass
  loss.backward()
  #parameters update
  with torch.no_grad():
    model.weights-= (learning_rate)*model.weights.grad
    model.bias -=(learning_rate)*model.bias.grad

  model.weights.grad.zero_()
  model.bias.grad.zero_()
  # printing loss
  print(f'epochs:{epoch+1}','Loss :',loss.item())
```

#Evaluation

```python
with torch.no_grad():
  y_pred=model.forward(X_test_tensor)
  y_pred=(y_pred>0.5).float()
acc=(y_pred==y_test_tensor).float().mean()
print(acc.item())
```
